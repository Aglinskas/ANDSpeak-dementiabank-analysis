```python
dfs = []
```

```python
files_dir = os.path.join(dir_root,'Data/DementiaBank/Pitt/dementia/')
files = [f for f in os.listdir(files_dir) if f.endswith('.mp3')]
paths = [os.path.join(files_dir,f) for f in files]

df = pd.DataFrame(paths,columns=['path'])
df['site'] = 'Pitt'
df['group'] = 'dementia'
df['filename'] = [val.split('/')[-1] for val in df['path'].values]
df['diag'] = 'D'
df['subName'] = ['Pitt_D_'+val[0:3] for val in df['filename'].values]
dfs.append(df)
```
