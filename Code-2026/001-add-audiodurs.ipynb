```python
import os
import pandas as pd
import numpy as np
import librosa
from tqdm import tqdm
```

```python
df = pd.read_csv('../DementiaBank-preprocessed2/file_list.csv')
df
```

```python
def get_duration(fn):
    y, sr = librosa.load(fn, sr=None)   # sr=None keeps original sampling rate
    duration_seconds = len(y) / sr
    duration_seconds = np.array(duration_seconds).round(2)
    return duration_seconds
```

```python
def get_size_mb(fn):
    size_bytes = os.path.getsize(fn)
    size_mb = size_bytes / (1024**2)
    size_mb = np.array(size_mb).round(2)
    return size_mb
```

```python
paths = df['path'].values
```

```python
sizes = [get_size_mb(fn) for fn in tqdm(paths)]
```

```python
durs = [get_duration(fn) for fn in tqdm(paths)]
```

```python
df['size_MB'] = sizes
df['dur_s'] = durs
df['dur_m'] = np.array(durs)/60
```

```python
df.to_csv('../DementiaBank-preprocessed2/file_list.csv')
```
