```python
import os
import pandas as pd
from tqdm import tqdm
```

```python
indir = '../DementiaBank-preprocessed2/04-speech-features/'
folders = [f for f in os.listdir(indir) if not f.startswith('.')]
folders.sort()
```

```python
ofdir = '../DementiaBank-preprocessed2/05-speech-features-merged/'
```

```python
for folder in tqdm(folders):
    files = os.listdir( os.path.join(indir,folder) )    
    if len(files)>0:
        ofn = files[0].split('_')[0]+'.csv'
        dfs = [pd.read_csv(os.path.join(indir,folder,file)) for file in files]
        pd.concat(dfs).mean(numeric_only=True).to_frame().T.to_csv(os.path.join(ofdir,ofn))
```
