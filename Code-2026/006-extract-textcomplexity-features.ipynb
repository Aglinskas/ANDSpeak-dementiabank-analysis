```python
import os 
import pandas as pd
import numpy as np

from calculate_text_complexity import calculate_text_complexity
```

```python
indir = '../DementiaBank-preprocessed2/02-patient-speaker-files/patient-text/'
```

```python
def get_text_lenght(fn):
    with open(fn) as f:
        a  = f.read()
    return len(a)
```

```python
files = [os.path.join(indir,f) for f in os.listdir(indir) if f.endswith('.txt')]
len(files)
text_length = [get_text_lenght(file) for file in files]
files = list(np.array(files)[np.array(text_length)>=300])
len(files)
```

```python
for f in np.arange(len(files)):
    fn = files[f]
    ofn = os.path.join(ofdir,fn.split('/')[-1].replace('.txt','.csv'))
    if not os.path.exists(ofn):
        calculate_text_complexity(fn,ofn)
```
