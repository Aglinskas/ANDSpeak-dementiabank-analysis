```python
indir = '../DementiaBank-preprocessed2/01-diarized-transcripts-v1/'
files = [f for f in os.listdir(indir) if f.endswith('.csv')]
len(files)a
```

```python
df = pd.read_csv(os.path.join(indir,files[0]))
```

```python
def merge_speaker_turns(df, max_gap=1.0):
    """
    Merge consecutive segments from same speaker
    if pause < max_gap seconds
    """

    merged = []
    current = df.iloc[0].copy()

    for i in range(1, len(df)):
        row = df.iloc[i]

        same_speaker = row.speaker == current.speaker
        small_pause = (row.start_sec - current.end_sec) < max_gap

        if same_speaker and small_pause:
            current.end_sec = row.end_sec
            current.duration_sec = current.end_sec - current.start_sec
            current.text += " " + row.text
        else:
            merged.append(current)
            current = row.copy()

    merged.append(current)

    return pd.DataFrame(merged)
```
