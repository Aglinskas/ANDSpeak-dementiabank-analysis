```python
import os
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
indir = '../DementiaBank-preprocessed2/01-diarized-transcripts-v1/'
files = [f for f in os.listdir(indir) if f.endswith('.csv')]
num_speaker = np.array([len(np.unique(pd.read_csv(os.path.join(indir,file))['speaker'].values)) for file in files])
files = np.array(files)[np.array(num_speaker>1)]
len(files)
```

```python
speaker_id = pd.read_csv('../DementiaBank-preprocessed2/patient_speaker_ids.csv')
speaker_id.sample(5)
```

```python
for f in tqdm(range(len(files))):
    file = speaker_id['file'].values[f]
    df = pd.read_csv(os.path.join(indir,file))
    idx=df['speaker'].values==speaker_id['patient_speaker'].values[f]
    speaker_df = df.iloc[idx]
    outdir = '../DementiaBank-preprocessed2/02-patient-speaker-files/patient-dfs/'
    speaker_df.to_csv(os.path.join(outdir,file))
```

```python
indir = '../DementiaBank-preprocessed2/02-patient-speaker-files/patient-dfs/'
files = [f for f in os.listdir(indir) if f.endswith('csv')]

for file in files:
    df = pd.read_csv(os.path.join(indir,file))
    transcript = (df.sort_values("start_sec")["text"].astype(str).str.strip().str.cat(sep=" "))
    
    ofdir = '../DementiaBank-preprocessed2/02-patient-speaker-files/patient-text/'
    ofn = os.path.join(ofdir,file.replace('.csv','.txt'))
    with open(ofn, "w", encoding="utf-8") as f:
        f.write(transcript)
```
