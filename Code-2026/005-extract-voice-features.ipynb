```python
import os
import pandas as pd
import numpy as np
from analyze_voice_features import analyze_voice_features
```

```python
indir = '../DementiaBank-preprocessed2/02-patient-speaker-files/patient-audio/'
```

```python
folders = os.listdir(indir)
folders = [f for f in folders if not f.startswith('.')]
len(folders)
```

```python
files = [[os.path.join(indir,folder,f) for f in os.listdir(os.path.join(indir,folder)) if f.endswith('.mp3')] for folder in folders]
files_stacked = np.hstack(files)

assert all([os.path.exists(val) for val in files_stacked]),'missing files'
```

```python
files_stacked = np.random.permutation(files_stacked)
```

```python
files_stacked.shape
```

```python
def process_one_file(fn):
    ofdir = '../DementiaBank-preprocessed2/04-speech-features/'
    subdir = os.path.join(ofdir,fn.split('/')[-2])
    if not os.path.exists(subdir):
        os.mkdir(subdir)

    ofn = os.path.join(ofdir,fn.split('/')[-2],fn.split('/')[-1].replace('.mp3','.csv'))
    if not os.path.exists(ofn):
        analyze_voice_features(fn,ofn)
```

```python
process_one_file(files_stacked[1])
```

```python
for file in files_stacked:
    try:
        process_one_file(file)
    except:
        pass
```
